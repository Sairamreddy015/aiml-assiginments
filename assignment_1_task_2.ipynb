```python
import matplotlib.pyplot as plt
import pandas as pd


people = ['kiran', 'arun', 'vijay', 'varun']
age =[25, 30, 35, 40]
height =[145, 151, 165, 173]
weight=[45, 55, 65, 75]

data=pd.DataFrame({
'name':people,
'Age':age ,
'Height':height,
'Weight':weight
})

plt.figure(figsize=(8, 6))
plt.scatter(data['Age'], data['Weight'], color='red')
plt.title('Scatter Plot of Age vs Weight')
plt.xlabel('Age')
plt.ylabel('Weight')

plt.figure(figsize=(8, 6))
plt.bar(data['name'], data['Height'], color='green')
plt.title('Bar Chart of Height for Each Person')
plt.xlabel('Name')
plt.ylabel('Height')

plt.figure(figsize=(8, 6))
plt.hist(data['Weight'], bins=10, color='orange', edgecolor='black')
plt.title('Histogram of Weight Distribution')
plt.xlabel('Weight')
plt.ylabel('Frequency')
```
